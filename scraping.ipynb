# IMDB Scraping

```python
from bs4 import BeautifulSoup
import requests
```

```python
html_text = requests.get('https://www.imdb.com/chart/top/?ref_=nv_mp_mv250').text
print(html_text)
```

```python
soup = BeautifulSoup(html_text, 'lxml')
soup
```

## Pegando os dados brutos sobre os filmes

<p>Importante lembrar que, os ttulos, os anos e as notas esto dispostos entre as tags, enquanto o diretor e quantidade de pessoas que votaram esto dispostas como atributos das tags.</p>
<p>Ento, vamos precisar trabalhar isso.</p>

### Diretor e titulos

```python
raw_titles = soup.find_all('td', class_='titleColumn')
```

```python
raw_titles[0]
```

```python
title = []
director = []
for element in raw_titles:
    for a in element.find_all('a'):
        title.append(a.text)
        director.append(a['title'])
        print(a['title'])
```

```python
director[0][:14]
```

```python
director_list = []
for i in range(0, 250):
    index = director[i].find('(')
    director_list.append(director[i][:index-1])
```

```python
director_list
```

### Ano, pontuao e quantidade de pessoas que votaram

```python
year = []
for t in raw_titles:
    for element in t.find_all('span'):
        year.append(element.text.replace('(', '').replace(')', ''))
year
```

```python
title = []
director = []
for element in raw_titles:
    for a in element.find_all('a'):
        title.append(a.text)
        director.append(a['title'])
        print(a['title'])
```

```python
raw_rating = soup.find_all('td', class_='ratingColumn imdbRating')
```

```python
rating = []
total_votes = []
for element in raw_rating:
    for strong in element.find_all('strong'):
        rating.append(strong.text)
        vote = strong['title'][strong['title'].find('on')+3:strong['title'].find('user')-1]
        total_votes.append(vote)
```

```python
rating
```

```python
total_votes
```

### Montar o dataframe e gerar um csv

<p>Antes de montar o dataframe precisamos trabalhar mais um pouco os dados, total_votes tem ',' separando milhar, coisa que no usamos para representar valores, devemos remove-las.</p>

```python
for i in range(0, 250): total_votes[i] = total_votes[i].replace(',', '')
```

```python
total_votes
```

```python
columns = ['filme', 'ano', 'nota', 'votos', 'diretor']
```

```python
import numpy as np
import pandas as pd
```

```python
values = [title, year, rating, total_votes, director_list]
```

```python
df = pd.DataFrame(values, index=columns).T
df
```

```python
df.to_csv('filmes.csv', sep=';', encoding='ISO-8859-1')
```
